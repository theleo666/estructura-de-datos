```python
class Examen:

    def __init__(self):
        self.f1=[]
        self.f2=[]
        self.f3=[]
        self.f4=[]
        self.axu=[]

    def normal(self, n=1):
        if n<4:
            self.f1.append(n)
            n+=1
            self.normal(n)
        elif n<7:
            self.f2.append(n)
            n+=1
            self.normal(n)
        elif n<10:
            self.f3.append(n)
            n+=1
            self.normal(n)
        elif n<13:
            self.f4.append(n)
            n+=1
            self.normal(n)
        else:
            print(self.f1)
            print(self.f2)
            print(self.f3)
            print(self.f4)
            print("\n")
    def inversa(self):
        self.f1.reverse()
        self.f2.reverse()
        self.f3.reverse()
        self.f4.reverse()
        print(self.f4)
        print(self.f3)
        print(self.f2)
        print(self.f1)
        print("\n")
        self.f1.reverse()
        self.f2.reverse()
        self.f3.reverse()
        self.f4.reverse()

    def caracol(self):
        if len(self.axu)>9:#123
            self.axu.pop()
            self.caracol()
        elif len(self.axu)>7:#45
            self.f2[2]=self.axu.pop()
            self.f3[2]=self.axu.pop()
            self.caracol()
        elif len(self.axu)>4:#678
            for j in range(0,3):
                self.f4.pop()
            for i in range(0,3):
                self.f4.append(self.axu.pop())
            self.f4.reverse()
            self.caracol()
        elif len(self.axu)>2: #910
            self.f3[0] = self.axu.pop()
            self.f2[0] = self.axu.pop()
            self.caracol()
        elif len(self.axu)>0: #1112
            self.f2[1] = self.axu.pop()
            self.f3[1] = self.axu.pop()
            self.caracol()
        else:
            print(self.f1)
            print(self.f2)
            print(self.f3)
            print(self.f4)

    def ayuda(self):
        self.axu.extend(self.f1)
        self.axu.extend(self.f2)
        self.axu.extend(self.f3)
        self.axu.extend(self.f4)
        self.axu.reverse()

a=Examen()
a.normal()
a.inversa()
a.ayuda()
a.caracol()
```
